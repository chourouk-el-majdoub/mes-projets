```python
#projet netflix partie 3
import pandas as pd

import seaborn as sns
import sqlalchemy as sql
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
donnees = pd.read_csv('/home/simplon/Documents/netflix_titles.csv')

#print (donnees.head())

donnees
```

```python
#Vous devez travailler sur un dataframe ou il n’y a pas de valeur manquante
categorie=donnees[['show_id','listed_in']].dropna()
categorie
```

```python
show_listed_in = pd.DataFrame(columns=['show_id', 'listed_in'])
for i in categorie.index:
    liste1 = categorie["listed_in"][i].split(", ")
    liste2 = [categorie["show_id"][i]] * len(liste1)
    df = pd.DataFrame({"show_id": liste2, "listed_in": liste1}, columns = ["show_id", "listed_in"])
    show_listed_in = show_listed_in.append(df)
show_listed_in
```

```python
#21
listed_in=show_listed_in["listed_in"].drop_duplicates()
listed_in = listed_in.reset_index(drop=True)
listed_in = listed_in.reset_index()
listed_in=listed_in.rename(columns={'index': 'listed_in_id'})

listed_in
```

```python
show_listed_in_id=pd.merge(show_listed_in,listed_in, on='listed_in', how='inner')
del show_listed_in_id['listed_in']
show_listed_in_id
```

```python
#22  show_director

directeur=donnees[['show_id','director']].dropna()
liste1=[]
liste2=[]

for i in range(directeur.shape[0]):
    liste1+=directeur.iloc[i,1].split(', ')
    liste2+=(directeur.iloc[i,0] for j in range(len(directeur.iloc[i,1].split(', '))))
show_director=pd.DataFrame({'show_id':liste2, 'director':liste1}) 
show_director
```

```python
print(show_director)
director=show_director['director'].drop_duplicates()
director=director.reset_index()
director=director.rename(columns={'index':'director_id'})
print(director)
show_director_id=pd.merge(show_director,director, on='director', how='inner')
del show_director_id['director']
show_director_id
```

```python
#22 show cast
acteur=donnees[['show_id','cast']].dropna()
liste1=[]
liste2=[]
for i in range(acteur.shape[0]):
    liste1+=acteur.iloc[i,1].split(', ')
    liste2+=(acteur.iloc[i,0] for j in range(len(acteur.iloc[i,1].split(', '))))
show_cast=pd.DataFrame({'show_id':liste2, 'cast':liste1}, columns=['show_id','cast'])  
print(show_cast)
cast=show_cast['cast'].drop_duplicates()
cast=cast.reset_index()
cast=cast.rename(columns={'index':'cast_id'})
print(cast)
show_cast_id=pd.merge(show_cast,cast, on='cast', how='inner')
del show_cast_id['cast']
show_cast_id
```

```python
#22 show country
pays=donnees[['show_id','country']].dropna()
print(pays.head(30))
liste1=[]
liste2=[]

for i in range(pays.shape[0]):
    liste1+=pays.iloc[i,1].split(', ')
    liste2+=(pays.iloc[i,0] for j in range(len(pays.iloc[i,1].split(', '))))
show_country=pd.DataFrame({'show_id':liste2, 'country':liste1})    
print(show_country )
country=show_country['country'].drop_duplicates()
country=country.reset_index()
country=country.rename(columns={'index':'country_id'})
print(country)
show_country_id=pd.merge(show_country,country, on='country', how='inner')
del show_country_id['country']
show_country_id
```
