```python
import pandas as pd
#pour les graphiques
#importation de la librairie
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('/home/simplon/Téléchargements/chomage_zone.xls',1,skiprows=5
)
print (data.head())
```

```python
p_emploi = pd.read_excel('/home/simplon/Téléchargements/pole_emploi.xlsx',0,skiprows=3
)
print (p_emploi)
```

```python
p_emploi.sort_values(by=['Code'])
data.sort_values(by=['REG'])
dataensemble=p_emploi.merge(data, how='right', left_on='Code', right_on='REG')
print(dataensemble)
```

```python
#En utilisant groupby() et mean() successivement calculer le taux de
#chômage moyen par régions pour chaque trimestres. Dans le groupby
#mettre as_index=‘False’. L’assigner à data_grp


data_grp = data.groupby(['REG','LIBREG'], as_index='False').mean()
print(data_grp)
```

```python

#Utiliser pandas.plot() pour tracer l’histogramme du taux de chômage du
#premier trimestre 2020. Les noms de régions doivent apparaitre en
#abscisse.


data_grp['2020-T1'].plot(x='LIBREG',kind='bar',y='REG')



```

```python
#En vous appuyant sur la documentation en ligne donné une taille correct
#au graphique, lui donner un titre et ajouter une grille en arrière-plan.

data_grp['2020-T1'].plot(x='LIBREG',kind='bar',y='REG',figsize=(20,7),title='taux de chômage',grid=True)
```

```python
#En repartant du jeux de données originel, isoler les colonnes 3 à 73. et assignées les à
#data2.
data2 = pd.read_excel('/home/simplon/Téléchargements/chomage_zone.xls',1,skiprows=5,usecols=range(3,73))
print (data2)
```

```python
#En utilisant groupby() et mean() successivement calculer le taux de chômage moyen par
#régions pour chaque trimestres. Cette fois ci ne pas préciser l’instruction as_index.
#L’assigner à data_grp2.
data_grp2 = data2.groupby(['LIBREG']).mean()
data_grp2
```

```python
#Transposer le dataframe
datatronspose=data_grp2.transpose()
datatronspose
```

```python
#Réindexer le dataframe avec reset_indadex().
data_t=datatronspose.reset_index()
data_t
```

```python
#Renommer le nom de la colonne ‘index’ en ‘periode’.
data_t.rename(columns={'index': 'periode'}, inplace=True)
data_t
```

```python
#Après avoir visualisé le contenu de data2, utiliser pandas.plot() pour tracer la ligne
#représentant l’évolution du taux de chômage pour la région ‘Provence-Alpes-Côte
#d'Azur’
data_t["Provence-Alpes-Côte d'Azur"].plot(x='periode',y="Provence-Alpes-Côte d'Azur",kind='line')
```

```python
#Répéter l’étape précédente sans préciser l’argument « y = ».
data_t["Provence-Alpes-Côte d'Azur"].plot(x='periode',kind='line')
```

```python
#Représenter l’évolution du taux de chômage pour la région ‘Provence-Alpes-Côte
#d'Azur’, ‘Occitanie’ et ‘Ile-de-France’ sur un même graphique.

data_t[["Provence-Alpes-Côte d'Azur",'Occitanie','Ile-de-France','periode']].plot(x='periode',kind='line',figsize=(15,6))
```
