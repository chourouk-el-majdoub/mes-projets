```python
#partie 2 Projet Netflix 
#12. Afficher les directeurs qui ont produit le plus de films/séries disponibles sur Netflix
directors=pd.Series(", ".join(donnees["director"].dropna()).split(", "))
directors.value_counts().head(10)
```

```python
#13. Voir si Jan Suter travaille souvent avec les mêmes acteurs
directors_notna=donnees[donnees['director'].notna()]
donnees_jan_suter=directors_notna[directors_notna['director'].str.contains("Jan Suter")]
donnees_jan_suter
jan_suter_cast=pd.Series(', '.join(donnees_jan_suter['cast'].dropna()).split(', '))
jan_suter_cast.value_counts().head(5)
```

```python
#14. Représenter les dix pays qui ont produit le plus de conte-nus disponibles sur Netflix, 
#avec le nombre de contenus parpays
donnees_pays=pd.Series(", ".join(donnees['country'].dropna()).split(", "))
donnees_pays=donnees_pays.value_counts().head(10)
donnees_pays
#suite 14
#donnees.loc[donnees['country'].isin(donnees_pays.index)]
dataframe_pays=donnees[donnees["country"].isin(donnees_pays.index)]
dataframe_pays
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=[20,10])
sns.countplot(x="country", data=dataframe_pays , order = dataframe_pays['country'].value_counts().keys())
plt.xticks(rotation=90)
plt.show()

```

```python
#15. Tracer un graphe à barres du nombre de films/séries parclassement de contenu (rating)
plt.figure(figsize=[20,10])
sns.countplot(x="rating", data=donnees , order = donnees['rating'].value_counts().keys())
plt.xticks(rotation=90)
plt.title("nombre de films/séries par classement de contenu (rating)")
plt.ylabel("nombre de films/séries")
plt.xlabel("Notation")
plt.show()


```

```python
#16. Afficher l’évolution du nombre de films/séries disponiblessur Netflix au cours du temps
data=donnees[donnees['date_added'].notna()]
date_year= pd.to_datetime(data['date_added']).dt.year
date_year
```

```python
donnees['year_added']=date_year
donnees
```

```python
data_group=donnees.groupby(['type','year_added']).size().reset_index(name='Count')
data_group
```

```python
plt.figure(figsize=[15,5])
plt.xticks(rotation=90)
sns.pointplot('year_added', 'Count', hue = 'type', data=data_group)
```

```python
#17. Afficher la distribution de la durée des films disponiblessur Netflix

movie = donnees[(donnees["type"]=="Movie")]
duree = pd.Series(movie['duration']).str.replace(" min", "").astype('int').sort_values(ascending=False)
sns.distplot(duree)
```
