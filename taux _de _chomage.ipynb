```python
#Charger le fichier « pole_emploi.xlsx » après l’avoir ouvert et visualiser et l’assigner à la
#variable « p_emploi ».
#• Vous ne pouvez pas utiliser les noms de colonnes contenus dans le fichier.
import pandas as pd
p_emploi = pd.read_excel('/home/simplon/Téléchargements/pole_emploi.xlsx',0,skiprows=3
)
print (p_emploi)
```

```python

#Joindre les data et p_emploi au seins d’un seul et unique dataframe en faisant
#correspondre les régions par leur codes.
#Utiliser pd.merge() pour réaliser cette jointure
p_emploi.sort_values(by=['Code'])
data.sort_values(by=['REG'])
dataensemble=p_emploi.merge(data, how='right', left_on='Code', right_on='REG')
print(dataensemble)






```

```python
#Créer une colonne quantile dans p_emploi qui prendra les valeurs [‘q25’, ‘q50’, ‘q75’,
#‘q100’] en fonction du quartile auquel appartient le nombre de pôle emploi. Pour se faire
#utiliser « pd.qcut() ».
dataensemble['quantile']=pd.qcut(dataensemble['Pôle emploi 2018'],q=4,labels=['q25', 'q50', 'q75','q100'])
print(dataensemble)
```

```python
#Générer un pivot de table montrant la moyenne du taux de chômage du trimestre 1 de
#l’année 2020 pour chaque régions et quantile du nombre de pôle emploi.
#Appeler la méthode pivot_table () sur le data frame
#Passer au paramètre « index » la valeur « LIBREG ».
#Passer au paramètre « columns » la valeur « quantile ».
#Passer au paramètre « values » la valeur «2020-T1 »
#Afficher les résultats.
pd.pivot_table(dataensemble,index=['LIBREG'],columns=['quantile'],values=['2020-T1'])
```
